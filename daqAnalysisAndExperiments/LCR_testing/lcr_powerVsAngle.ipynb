```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



plt.rc('font', family='serif')
plt.rc('xtick', labelsize=14)
plt.rc('ytick', labelsize=14)
plt.rc('axes', titlesize=20, labelsize=20)
plt.rc('legend', fontsize=12)
plt.rc('figure', titlesize=18)
plt.rcParams["date.autoformatter.minute"] = "%d %H:%M:%S"

%matplotlib qt
```

```python
file_path = './data.txt'
df = pd.read_csv(file_path)
df.head()
```

```python
std_values = df.drop('Angle', axis=1).std()
df.loc['STD'] = std_values
df.tail()
```

```python
plt.close('all')
plt.figure()
angles = df['Angle']
for i in df.columns[1:7]:
    std = df[i].std()
    plt.plot(angles, 10*np.log10(df[i]), label = f'{i}. STD = {std:.2f}[lin. W] ' )
plt.xlabel('Angle [deg]')
plt.ylabel('Power [dBm]')
plt.tight_layout()
plt.legend()

plt.figure()
angles = df['Angle']
for i in df.columns[6:]:
    std = df[i].std()
    plt.plot(angles, 10*np.log10(df[i]), label = f'{i}. STD = {std:.5f}[lin. W] ' )
plt.xlabel('Angle [deg]')
plt.ylabel('Power [dBm]')
plt.tight_layout()
plt.legend()
```
